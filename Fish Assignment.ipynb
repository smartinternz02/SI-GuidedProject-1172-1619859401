```python
x = dataset.iloc[:,[0,2,3,4,5,6]].values
```

```python
y = dataset.iloc[:,1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()

ct=ColumnTransformer([('oh',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10000)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
ypred = mlr.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(ypred,y_test)
```

```python
accuracy
```
